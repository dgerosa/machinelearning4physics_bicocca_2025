## The expansion of the Universe

This project investigates a well-known problem in Cosmology (a branch of astrophysics that studies the origin, structure, evolution, and ultimate fate of the Universe as a whole): **the expansion of the Universe**! 


In 1920, Hubble, from the observation of a special kind of stars known as Cepheid was able to determine their distance and those of the galaxies where they belong. For these galaxies, he was also able to measure the velocity. From these measurments, Hubble found the existence of a linear relationship between the velocity of the galaxies and their distance from us. [[Hubble 1929](https://www.pnas.org/doi/10.1073/pnas.15.3.168)]

![](https://www.e-education.psu.edu/astro801/sites/www.e-education.psu.edu.astro801/files/image/hubbleslaw_plot.jpg)

That is, things at larger distances are moving faster. This means the Universe is expanding! 



Hubble formulated that the velocity of the galaxies, $v$, can be related to the distance, $d$, through the following relationship: $v = H_0 d$, where $H_0$ is known as the Hubble constant.

But, how do we measure distances and velocities in astronomy?

- **Distances**: In real life, they are measured with some physical instrument (e.g., a ruler). However, in astronomy, usually we have to deal with objects that are very far apart (the distance to the Andromeda galaxy, the closest galaxy to the Milky Way, is of the order of $10^{19}$ km), and we have to use a different approach: absolute and apparent brightness. Suppose we have a source of known (i.e., intrinsic) brightness, e.g., a light bulb. If we move the source far away from us and we measure the brightness, since we know the real, intrinsic one, we can estimate the distance at which the source has been moved. We can use this idea also for some astrophysical objects, but we need to know the intrinsic brightness (not an easy task). For some objects, we know it, for example, Cepheid stars and supernovas. We can infer the distance by considering the difference between the intrinsic magnitude (or brightness) and apparent (measured) magnitude (or brightness).


- **Velocities**: When a source emitting light or sound is moving, the wavelength we measure is different compared to the case when the source is still. This is the Doppler effect (e.g., ambulance example in any high school physics book). Specifically, how the wave is modified depends on whether the source is moving away from us (redshifted) or toward us (blue-shifted). By measuring the amount of shift the wave has experienced, we can measure the velocity. This can be exploited with specific electromagnetic emissions coming from the galaxies. From the redshift, we can infer the velocity.

Hubble's linear relationship between the velocity at which galaxies are running away from us and their distance holds just for the local Universe, i.e., for galaxies very near to us. This is due to the fact that variable stars, like Cepheids, are too hard to see when they are very far away.

At high redshifts, cosmology is done with a specific kind of exploding stars called *Type Ia Supernovae*. As for Cepheids, for this kind of source we know the intrinsic brightness, and once we measure their apparent brightness, we can infer the distance.

We know the intrinsic brightness of Type Ia Supernovae since we understand the mechanism by which they are produced.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/c/ce/Progenitor_IA_supernova.svg/1280px-Progenitor_IA_supernova.svg.png)



When they started using Type 1A Supernova for cosmology, they found another mindblowing surprise!

This is a [Nobel prize](https://www.nobelprize.org/prizes/physics/2011/riess/facts/) plot by [Riess et al](https://iopscience.iop.org/article/10.1086/300499):



Here, on the y-axis (top panel), $\mu = m - M$ (the difference between the apparent and intrinsic magnitude) is plotted, which can be directly interpreted as a distance, and on the x-axis, there is $z$, the redshift, which can be interpreted as the velocity. As you can see, at small distances and velocities, the relationship is still linear, but at high redshift, it deviates!
The non-linearity at high redshift implies that the Universe is not only expanding, but it is accelerating! That is, the expansion in the past was slower than the expansion today.

Following the more established and accepted cosmological model of the Universe, [$\Lambda CDM$ ](https://en.wikipedia.org/wiki/Lambda-CDM_model):
How the Universe is expanding (at a constant or accelerated rate) depends on the content of the Universe itself. From the simplest to the more sophisticated observations, we know that the Universe is made up of *ordinary matter* (atoms, particles, etc.), *light* (photons), and *dark matter* (a special kind of matter that does not interact with ordinary matter or light, but interacts through gravity. It is essential to explain the properties of galaxies and is a very active field of research). Each of these components is associated with a density:
- $\Omega_{\rm OM}$ the density of ordinary matter in the Universe
- $\Omega_{\rm ph}$ the density of photons energy (light) in the Universe
- $\Omega_{\rm DM}$ the density of dark matter in the Universe


However, just these three components are not enough to explain the accelerating expansion of the Universe. For this reason, another component, known as *dark energy*, has been introduced. We have no idea how to interpret this quantity, apart from the fact that we need it to explain the accelerated expansion. To this energy, we associate the density $\Omega_{\Lambda}$.

These $\Omega$ parameters also determine the shape of the Universe. We believe that we live in a flat Universe, which implies:

$$\Omega_{\rm OM}+\Omega_{\rm ph}+\Omega_{\rm DM}+\Omega_{\rm \Lambda}=1$$

where $\Omega_{\rm OM}$ and $\Omega_{\rm ph}$ are very small compared to $\Omega_{\rm DM}$ (dark matter) and $\Omega_{\rm \Lambda}$ (dark energy), and thus can be *ignored*!


In $\Lambda CDM$ , the distance modulus is given by (see e.g. [Hogg 2010](https://arxiv.org/abs/astro-ph/9905116)):

$$\mu = 5 \log \left( \frac{c/H_0} {10 {\rm pc}}(1+z)\int_0^z \frac{dz'}{\sqrt{\Omega_{\rm DM} (1+z')^3+\Omega_\Lambda}} \right) $$


*Reminder:* the parameters of the model are
- $H_0$, the Hubble constant
- $\Omega_{D\rm M}$ the density of dark matter in the Universe
- $\Omega_\Lambda$ the density of dark energy in the Universe
- (the density of ordinary matter is so small that can be safely neglected). 
- (we are also neglecting the energy of photons, i.e. light)

*Notes*:
- $c$ is the expression above is the speed of light 
- pc in the expression above stands for parsec with 1 pc = 3.086 x 10^16 meters.
- Astronomers like weird units. The Hubble constant has unit of 1/s but it's typically quoted in (km/s)/Mpc. 
- The Omegas are dimensionless.
